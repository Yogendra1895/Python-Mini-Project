#### 20. Create a bar chart for all games. 

```python
plt.figure(figsize=(10,6))

All_Genre = df['Genre'].value_counts()
All_Genre.plot(kind='bar', color="blue",edgecolor="black")

plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.title('Number of Games per Genre')
plt.xticks(rotation=90)
plt.show()

```

#### 21. Which developer has most games sold:- top 20 developer.

```python
top_developers = (
    df.groupby("Developer")["Global_Sales"].sum().sort_values(ascending=False).head(20).reset_index()
)
print(top_developers)
```

#### 22. Plot the bar chart for the above top 20 developers.

```python
plt.figure(figsize=(10,5))

Top_20_Developers = df.groupby('Developer')['Global_Sales'].sum()

Top_20_Developers.plot(kind='bar',color='green', edgecolor='black')

plt.xlabel('Developer')
plt.ylabel('Total Global Sales (in millions)')
plt.title('Top 20 Developers by Global Sales')
plt.xticks(rotation=90)
plt.show()
```
